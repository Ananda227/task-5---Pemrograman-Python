## Import library Pandas, Matplotlib, dan Seaborn

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(color_codes=True)

import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("/Kuliah/python/task5/Dataset/iris.csv")
df.head()
```

## Membuat Scatter Plot dari kolom SepalLenghtCm dan SepalWidthCm

```python
df.plot(kind="scatter", x="SepalLengthCm", y="SepalWidthCm")
```

## Menggunakan Seborn

```python
sns.jointplot(x="SepalLengthCm", y="SepalWidthCm", data=df, size=5)
```

## Mewarnai sebaran Species

```python
# keyword 'size' gak terdefinisi jadi saya menggunakan  aspect
sns.FacetGrid(df, hue="Species", palette="husl", aspect=2).\
map(plt.scatter, "SepalLengthCm", "SepalWidthCm").\
add_legend()
```

## Plot dengan warna yang berbeda

```python
sns.FacetGrid(df, hue="Species", palette="hls", aspect = 2).map(plt.scatter, "SepalLengthCm", "SepalWidthCm").add_legend()

sns.FacetGrid(df, hue="Species", palette="Set2", aspect = 2).map(plt.scatter, "SepalLengthCm", "SepalWidthCm").add_legend()

sns.FacetGrid(df, hue="Species", palette="cubehelix", aspect = 2).map(plt.scatter, "SepalLengthCm", "SepalWidthCm").add_legend()
```

## Pairplot


dibawah ini kolom id dihapus dikarenakan tidak memiliki korelasi dengan kolom lain
